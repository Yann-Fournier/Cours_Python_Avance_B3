### Exercice 3 Disponibilité des vélibs
---

1. Importer les données sous la forme d’un dataFrame
- velib_t1.txt  - avec les données des stations à un instant t
- velib_t2.txt  - avec les données des stations à un instant t+1

```python
with open("velib_t1.txt", "r", encoding="utf-8") as f:
    contenu_json = json.loads(f.read())

df_velib_t1 = pd.DataFrame(contenu_json)
print(df_velib_t1.shape)
df_velib_t1.head()
```

```python
with open("velib_t2.txt", "r", encoding="utf-8") as f:
    contenu_json = json.loads(f.read())

df_velib_t2= pd.DataFrame(contenu_json)
print(df_velib_t2.shape)
df_velib_t2.head()
```

2. Représenter la localisation des stations vélib dans Paris
représenter les stations en fonction du nombre de places avec
un gradient

```python
def create_map(df: pd.DataFrame) -> folium.Map:
    """
    Crée une carte Folium avec des cercles colorés représentant les stations de vélos.
    Les couleurs sont basées sur le nombre de places disponibles dans chaque station.
    """
    # Normaliser le nombre de places pour le gradient de couleurs
    min_places = df['bike_stands'].min()
    max_places = df['bike_stands'].max()

    def get_color(value, min_val, max_val):
        norm = (value - min_val) / (max_val - min_val)  # Normalisation entre 0 et 1
        cmap = mcolors.LinearSegmentedColormap.from_list("gradient", ["green", "yellow", "red"])
        return mcolors.to_hex(cmap(norm))  # Convertit en couleur hexadécimale

    # Créer la carte centrée sur Paris
    m = folium.Map(location=[48.8566, 2.3522], zoom_start=12)

    # Ajouter les stations avec des cercles colorés
    for _, row in df.iterrows():
        color = get_color(row['bike_stands'], min_places, max_places)
        folium.CircleMarker(
            location=[row['position']['lat'], row['position']['lng']],
            radius=5,  # Taille du point
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=f"Station: {row['name']}<br>Places: {row['bike_stands']}"
        ).add_to(m)

    return m # Afficher la carte
```

```python
# On créer une carte avec les stations de vélos à Paris
create_map(df_velib_t1[df_velib_t1['address'].str.contains('PARIS')].copy())
# Temps 0.8s
```

```python
# On créer une carte avec les stations de vélos à Paris
create_map(df_velib_t2[df_velib_t2['address'].str.contains('PARIS')].copy())
# Temps 0.8s
```

3. Comparer pour une station donnée l’évolution de la disponibilité (en fusionnant les deux bases t et t+1)

- représenter les stations qui ont connu une évolution
significative (plus de 5 changements) avec un gradient de
couleurs

```python
df_merge_velib = pd.merge(df_velib_t1, df_velib_t2, on='name', suffixes=('', '_t2'))
df_merge_velib['Changement'] = df_merge_velib['available_bikes_t2'] - df_merge_velib['available_bikes'] # ajout colonne changement
df_merge_velib['Changement abs'] =  df_merge_velib['Changement'].abs() # valeur absolue du changement
print(df_merge_velib.shape)
df_merge_velib.head()
```

```python
# On créer une carte avec les stations avec un changement de vélos supérieur à 5
create_map(df_merge_velib[df_merge_velib['Changement abs'] > 5].copy())
# Temps 0.0s
```
